##  Import Libraries

```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Read Image
im=cv2.imread("wifire/sd-3layers.jpg")
print(type(im))
print(im.shape)
```

```python
# Display image using CV2
cv2.imshow("image",im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Images are red in BGR mode unlike RGB we read earlier.
print(im[2000,2900,:]) #Near Green/Blue
print(im[2500,3600,:]) #Near Red
```

```python
# Change it to RGB
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
# im_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("image",im_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im_rgb)
```

```python
#Resize the image
resized = cv2.resize(im, (1000, 1000))
cv2.imshow("Fixed Resizing", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Maintain aspect ratio
w=1000
r=w/im.shape[1] #calculte ratio in which width is decreased.
h=int(im.shape[0]*r) #calculate new height

resized = cv2.resize(im, (w, h))
cv2.imshow("Fixed Resizing", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(resized)
```

```python
#Blur and image
im1=cv2.imread("woman.jpg")
h=int(im1.shape[0]*.5)
w=int(im1.shape[1]*.5)
im1=cv2.resize(im1,(w,h))
# blurred = cv2.GaussianBlur(resized, (11, 11), 0)
cv2.imshow("Woman", im1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Blur and image
blurred = cv2.GaussianBlur(im1, (11, 11), 0)
cv2.imshow("Blurred", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Convert to gray scale
gray = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Detect Edge
canny = cv2.Canny(im1,30,250)
cv2.imshow("Canny", canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Draw Rectangle
output = im1.copy()
cv2.rectangle(output, (320, 60), (420, 160), (0, 0, 255), 2)
cv2.circle(output, (300, 150), 20, (255, 0, 0), -1)
cv2.putText(output, "OpenCV + Jurassic Park!!!", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
cv2.imshow("Rectangle", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
